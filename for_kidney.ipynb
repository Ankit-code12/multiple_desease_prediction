```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("kidney_disease.csv")
df.shape
df.info()
df.describe()
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
(df==0).sum()
```

```python
num_cols=df.select_dtypes(include='number').columns
for col in num_cols:
    df[col].fillna(df[col].median(),inplace=True)
```

```python
cat_cols=df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col].fillna(df[col].mode(),inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
for col in cat_cols:
    df[col]=le.fit_transform(df[col])
```

```python
X=df.drop('classification',axis=1)
Y=df['classification']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

kidney_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    random_state=42
)

kidney_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = kidney_model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import joblib
import os

os.makedirs("models", exist_ok=True)

joblib.dump(kidney_model, "models/kidney_rf_model.pkl")
joblib.dump(scaler, "models/kidney_scaler.pkl")
```

```python
new_patient = [
    45, 80, 1.020, 1, 0, 1, 1, 0, 0, 120, 
    1.2, 135, 4.5, 14, 15, 1, 0, 1, 1, 1, 1, 0, 1,0,1
]
  # Example values

new_patient_scaled = scaler.transform([new_patient])
prediction = kidney_model.predict(new_patient_scaled)

if prediction[0] == 1:
    print("Kidney Disease Detected")
else:
    print("No Kidney Disease")
```
