```python
import pandas as pd
```

```python
df=pd.read_csv("diabetes.csv")
df.head()
df.info()
```

```python

cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in cols:
    df[col]=df[col].replace(0,df[col].median())
```

```python
df.head()
(df==0).sum()
```

```python
X=df.drop('Outcome',axis=1)
Y=df['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    random_state=42
)

rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)
y_pred[:10]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
import joblib
import os

os.makedirs("models", exist_ok=True)

joblib.dump(rf_model, "models/diabetes_rf_model.pkl")
joblib.dump(scaler, "models/diabetes_scaler.pkl")

```

```python
import joblib

model = joblib.load("models/diabetes_rf_model.pkl")
scaler = joblib.load("models/diabetes_scaler.pkl")

patient_data = [2, 120, 70, 20, 85, 28.5, 0.45, 35]  # for one patient 

scaled_data = scaler.transform([patient_data])

prediction = model.predict(scaled_data)
print("pridiction :",prediction)

if prediction[0] == 1:
    print("Diabetes Detected")
else:
    print("No Diabetes")
```

```python
patients_data = [
    [2, 120, 70, 20, 85, 28.5, 0.45, 35],   # patient 1
    [1, 140, 80, 30, 90, 32.0, 0.5, 40],    # patient 2
    [3, 100, 60, 15, 70, 25.0, 0.4, 30]     # patient 3
]

scaled_data = scaler.transform(patients_data)
predictions = model.predict(scaled_data)
print(predictions)

```
